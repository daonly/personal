```python
# 2. 패키지 임포트
import pandas as pd
import bambi as bmb
import arviz as az
import matplotlib.pyplot as plt


# 3. 데이터 준비: stanIAw, initDist, absChan만 추출
from google.colab import files
uploaded = files.upload()
df = pd.read_excel("All.xlsx")
df_filtered = df[['stanIAw', 'initDist', 'absChan']].dropna()

# 4. 모델 1: stanIAw → initDist
model1 = bmb.Model("initDist ~ stanIAw", data=df_filtered)
results1 = model1.fit(draws=2000, chains=4)

# 5. 모델 2: stanIAw → absChan
model2 = bmb.Model("absChan ~ stanIAw", data=df_filtered)
results2 = model2.fit(draws=2000, chains=4)

# 6. 결과 출력 및 요약
print("StanIAw → InitDist")
az.summary(results1, var_names=["stanIAw"])

print("\nStanIAw → AbsChan")
az.summary(results2, var_names=["stanIAw"])

# 7. 사후분포 시각화 (선택)
az.plot_posterior(results1, var_names=["stanIAw"])
plt.title("Effect of stanIAw on initDist")
plt.show()

az.plot_posterior(results2, var_names=["stanIAw"])
plt.title("Effect of stanIAw on absChan")
plt.show()
```
