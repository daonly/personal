## SE 개인별, 조건별 event related plotting

```python
### preparation

import os
import glob
import matplotlib.pyplot as plt
import numpy
import pandas as pd

def cd(path):
    os.chdir(path)
```

```python
### set path 

basepath = r'C:\newdeal\EDA analysis\EDA_data'

taskname = 'SEf'  ## 변경 필!
datapath = os.path.join(basepath,'SEf_FBOnset-3FBEvents')  ## 변경 필!
folders = glob.glob(datapath+'\*.csv')
savepath = os.path.join(datapath,'plot')
```

```python
### get data from csv & plotting

x = [-1,-0.75,-0.5,-0.25,0,0.25,0.5,0.75,1,1.25,1.5,1.75,2,2.25,2.5,2.75,3,3.25,3.5,3.75,4,4.25,4.5,4.75,5]

for i in range(0,len(folders)):
    subname = folders[i].split('\\')[-1].split('_')[0]
    condname = folders[i].split('\\')[-1].split('_')[-1].split('.')[0]
    csv = pd.read_csv(folders[i],names=['-1','-0.75','-0.5','-0.25','0','0.25','0.5','0.75','1','1.25','1.5','1.75','2','2.25','2.5','2.75','3','3.25','3.5','3.75','4','4.25','4.5','4.75','5'])
    csv = csv.transpose()
    for j in range(0,len(csv.columns)):
        trial = csv[j].values.tolist()
        plt.plot(x,trial)
        plt.xlabel("time series")
        plt.ylabel("phasicData")
        plt.savefig(savepath+f'\{subname}_{taskname}_{condname}.png')
    plt.close()
```
