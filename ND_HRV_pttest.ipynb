```python
### preparation

import os
import glob
import numpy
import pandas as pd
```

```python
### set path 

basepath = r'C:\newdeal\HRV analysis'

taskname = 'HRV'  ## 변경 필!
datapath = os.path.join(basepath,'raw')  ## 변경 필!
folders = glob.glob(datapath+'\*resampled.physioData')
savepath = os.path.join(basepath,'analysis')
```

```python
### get sub info
file_list=[]
for w in folders:
    file_list.append(w.split('\\')[-1].split('.')[0])

file_list.remove('CMJ1704_HRV_ECG_20220922140615_resampled')
file_list.remove('PJY5851_HRV_ECG_20220921131146_resampled')
file_list.remove('PSJ7390_HRV_ECG_20220829130927_resampled')

### organize sub info
sub_list=[]
for w in file_list:
    sub_list.append(w.split('_')[0])

for i in range(0,len(sub_list)):
    if sub_list.count(file_list[i].split('_')[0]) == 1: #여기서 나오는 오류는 무시 ㄱㄴ!
        del file_list[i]
```

### 아래의 세 코드 중 원하는 domain 하나만 돌릴 것!

```python
### fill the table for time-domain indices
df = pd.read_excel(datapath+'\PhysioEpochResults_20230214T152823_set.xlsx')
df = df[['dataSource','RMSSD_detrended','IBI_std_detrended','pNN20_detrended','pNN50_detrended']]
df.set_index('dataSource', inplace = True)
df
```

```python
### fill the table for freq-domain indices
df = pd.read_excel(datapath+'\PhysioEpochResults_20230214T152823_set.xlsx')
df = df[['dataSource','VLF_powerPercent','VLF_power','VLF_psdPeak','VLF_psdPeakFreq','LF_powerPercent','LF_power','LF_psdPeak','LF_psdPeakFreq','HF_powerPercent','HF_power','HF_psdPeak','HF_psdPeakFreq','LF_HF_ratio']]
df.set_index('dataSource', inplace = True)
df
```

```python
### fill the table for non-linear-domain indices
df = pd.read_excel(datapath+'\PhysioEpochResults_20230214T152823_set.xlsx')
df = df[['dataSource','poincare_SD1','poincare_SD2','poincare_SD2_SD1_ratio']]
df.set_index('dataSource', inplace = True)
df
```

### 여기서부터 본격적으로 통계 준비!

```python
### split into each dataset

c_list = list(df.columns)

for i in range(0,len(c_list)):
    globals()[c_list[i]+'_1'] = []
    globals()[c_list[i]+'_2'] = []

for j in range(0,len(file_list),2):
    day1 = df.loc[[file_list[j]]]
    day2 = df.loc[[file_list[j+1]]]
    
    for i in range(0,len(c_list)):
        globals()[c_list[i]+'_1'].append(day1.values.tolist()[0][i])
        globals()[c_list[i]+'_2'].append(day2.values.tolist()[0][i])
```

```python
### conduct paired t-tests

import scipy.stats

for i in range(0,len(c_list)):
    res = f"{c_list[i]}: {scipy.stats.ttest_rel(globals()[c_list[i]+'_1'],globals()[c_list[i]+'_2'])}"
    print(res)
```
